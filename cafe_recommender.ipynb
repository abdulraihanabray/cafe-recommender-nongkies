```python
# Calculate similarity matrix
similarity_matrix = cosine_similarity(count_matrix, count_matrix)
features = ['start_price', 'wifi_status', 'wifi_speed',]
feature_matrix = pd.get_dummies(cafe_list[features])

# Recommend function
def recommend_cafe(cafe_name, similarity_matrix, n=3):
    idx = cafe_list[cafe_list['name'] == cafe_name].index[0]
    similar_cafes = sorted(list(enumerate(similarity_matrix[idx])), key=lambda x: x[1], reverse=True)[1:n+1]
    recommended_cafes = [cafe_list.iloc[i[0]]['name'] for i in similar_cafes]
    return recommended_cafes
```

```python
# Example usage
cafe_name = 'Morgy Coffee'
recommendations = recommend_cafe(cafe_name, similarity_matrix)
print("Recommended cafes for {}: {}".format(cafe_name, recommendations))
```
