```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset.csv', sep=",")
```

```python
df.head()
```

```python
df['lokasi'].value_counts().head()
```

```python
df['lokasi'][54].split(", ")
```

```python
# Determine the maximum number of splits among all rows
max_splits = df['lokasi'].str.count(',').max() + 1

# Split the 'lokasi' column into multiple columns
split_columns = [f'lokasi_part_{i+1}' for i in range(max_splits)]
df[split_columns] = df['lokasi'].str.split(", ", expand=True)
```

```python
df.head(2)
```

```python
# Identify rows where lokasi_part_6 and lokasi_part_7 are not None
rows_to_delete = df[(df['lokasi_part_6'].notnull()) | (df['lokasi_part_7'].notnull())].index

# Delete rows from the DataFrame
df = df.drop(rows_to_delete)

# Reset the index
df = df.reset_index(drop=True)
```

```python
df.head(2)
```

```python
df['wifi'].value_counts()
```

```python
# Convert values based on conditions
df['wifi'] = np.where((df['wifi'] == 'Wi-Fi Not Available') | (df['wifi'] == 'Unknown (Due to Wi-Fi Issue)'), 'Unavailable', df['wifi'])

# Split the 'wifi' column based on "-"
split_wifi = df['wifi'].str.split(" - ", expand=True)

# Create new columns for each part
df['wifi_status'] = split_wifi[0]
df['wifi_speed'] = split_wifi[1]

# Extract only numbers from the 'amount' column
df['range_harga'] = df['range_harga'].str.extract('(\d+)')
```

```python
df.head(2)
```

```python
# Extract only numbers from the 'address' column
df['postal_code'] = df['lokasi_part_5'].str.extract('(\d+)')

# Remove numbers from the 'address' column
df['lokasi_part_5'] = df['lokasi_part_5'].str.replace('\d+', '')

df.head(2)
```

```python
df.drop(columns=['id', 'lokasi', 'wifi', 'lokasi_part_6', 'lokasi_part_7'], inplace=True)
df.head(2)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df = df[['nama', 'kategori', 'range_harga', 'wifi_status', 'wifi_speed', 'kesimpulan',
        'point to consider', 'lokasi_part_1', 'lokasi_part_2', 'lokasi_part_3',
        'lokasi_part_4', 'lokasi_part_5', 'postal_code']]

df.rename(columns={'nama':'name', 'kategori':'category', 'range_harga':'start_price', 'kesimpulan':'conclusion',
                   'lokasi_part_1':'street', 'lokasi_part_2':'urban_village', 'lokasi_part_3':'district', 'point to consider':'point_to_consider',
                   'lokasi_part_4':'regency', 'lokasi_part_5':'province'}, inplace=True)
```

```python
df['regency'] = df['regency'].replace('Bandung City', 'Kota Bandung')
df['district'] = df['district'].str.replace('Kec. ', '').str.replace('Kecamatan ', '')
```

```python
df.to_csv('dataset_v2.csv', index=False)
```
